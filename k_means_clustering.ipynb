```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Đọc latent embedding ---
latent = pd.read_csv("luad_omics_latent.csv", index_col=0)
print(f"Shape latent: {latent.shape}")
```

```python
scores = []
K_range = list(range(2, 7))
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    labels = kmeans.fit_predict(latent)
    score = silhouette_score(latent, labels)
    scores.append(score)
    print(f"k={k}, silhouette={score:.3f}")
```

```python
# --- Phân cụm với k tối ưu ---
k_opt = K_range[scores.index(max(scores))]
print(f"Optimal k: {k_opt}")
kmeans = KMeans(n_clusters=k_opt, random_state=42, n_init=20)
labels = kmeans.fit_predict(latent) 
# print(labels)
latent['cluster'] = labels
latent.to_csv(f"latent_kmeans_cluster_k{k_opt}.csv")
print("Cluster sizes:", latent['cluster'].value_counts())
```
